````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2, 4, 5, 4, 5])
coefficients = np.polyfit(x_data, y_data, 1)
fit_function = np.poly1d(coefficients)
x_fit = np.linspace(min(x_data), max(x_data), 100)
y_fit = fit_function(x_fit)
plt.scatter(x_data, y_data, label='Data')
plt.plot(x_fit, y_fit, label='Fitted Curve', color='red')
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
````

````python
# Importing necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Generating some example data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Creating a linear regression model
model = LinearRegression()

# Training the model
model.fit(X, y)

# Making predictions
X_new = np.array([[0], [2]])
y_pred = model.predict(X_new)

# Plotting the data and the regression line
plt.scatter(X, y, color='blue', label='Original data')
plt.plot(X_new, y_pred, color='red', linewidth=2, label='Regression line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

````

````python
# Importing necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

# Generating some example data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + 1.5 * X**2 + np.random.randn(100, 1)

# Transforming the input features to include polynomial terms up to degree 2
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)

# Creating a linear regression model
model = LinearRegression()

# Training the model with polynomial features
model.fit(X_poly, y)

# Generating data for plotting
X_new = np.linspace(0, 2, 100).reshape(-1, 1)
X_new_poly = poly_features.transform(X_new)
y_pred = model.predict(X_new_poly)

# Plotting the data and the polynomial regression curve
plt.scatter(X, y, color='blue', label='Original data')
plt.plot(X_new, y_pred, color='red', linewidth=2, label='Polynomial Regression (degree=2)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Generating some example data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 3 * np.exp(0.5 * X) + np.random.randn(100, 1)

# Performing exponential regression
coefficients = np.polyfit(X.flatten(), np.log(y.flatten()), deg=1)
a, b = coefficients
a = np.exp(a)  # Transforming back from the logarithmic scale

# Generating data for plotting
X_new = np.linspace(0, 2, 100)
y_pred = a * np.exp(b * X_new)

# Plotting the data and the exponential regression curve
plt.scatter(X, y, color='blue', label='Original data')
plt.plot(X_new, y_pred, color='red', linewidth=2, label='Exponential Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randint(1, 21, 100)
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = (data < lower_bound) | (data > upper_bound)
plt.scatter(range(len(data)), data, color='blue', label='Data points')
plt.scatter(np.where(outliers)[0], data[outliers], color='red', label='Outliers')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()
````

````python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]


data = list(zip(x, y))
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
````

````python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data, true_labels = make_blobs(n_samples=30, centers=3, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)
predicted_labels = kmeans.fit_predict(data)
final_centroids = kmeans.cluster_centers_

plt.scatter(data[:, 0], data[:, 1], label='Predicted')
plt.scatter(final_centroids[:, 0], final_centroids[:, 1],marker='x', label='Centroids')
plt.title('K-means Clustering')
plt.legend()
plt.show()
````
