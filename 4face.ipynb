## HANDLING OF MISSING DATA -3a
 create a sample data frame and some missing values

```python
import numpy as np
import pandas as pd
df=pd.DataFrame({"name":["alice","bob","getha"],
                   "age":[10,20,30],
                   "gender":["m","f","m"]})

select_row=df.loc[0]
select_col=df.loc[:,"name"]

print("selectrow:\n",select_row)
print("selct col:\n",select_col)

df.loc[[0,1],"age"]=np.nan
print("dataframe:\n",df)
```

## HANDLING OF MISSING DATA -3b
perform drop a rows and columns that having a missing value

```python
df=pd.DataFrame({"name":["alice","bob","getha"],
                   "age":[np.nan,np.nan,30],
                   "gender":["m","f","m"]})

all=df.dropna()
row=df.dropna(thresh=3)
col=df.dropna(axis=1)

print(all)
print(row)
print(col)
```

## HANDLING OF MISSING DATA -3c
 perform drop based on a particular subset of columns

```python
import pandas as pd
import numpy as np
df=pd.DataFrame({"name":["alice","bob","getha"],
                   "age":[np.nan,np.nan,30],
                   "gender":["m","f","m"]})
specific_col = df.drop(columns=["age"])
print(specific_col)
```

## HANDLING OF MISSING DATA -3d
 missing values with constant and aggregated value

```python
df=pd.DataFrame({"name":["alice","bob","getha"],
                   "age":[np.nan,np.nan,30],
                   "gender":["m","f","m"]})

fillconstant=df.fillna(10)
fillmean=df.fillna(df.mean(numeric_only=True))
fillmode=df.fillna(df.mode(numeric_only=True))
fillmedian=df.fillna(df.median(numeric_only=True))

print(fillconstant)
print(fillmean)
print(fillmode)
print(fillmedian)
```

## HANDLING OF MISSING DATA -3e
 the missing values with the previous/next value

```python
import numpy as np
import pandas as pd
df=pd.DataFrame({"name":["alice","bob","getha"],
                   "age":[10,np.nan,30],
                   "gender":["m","f","m"]})
prev=df.fillna(method='pad')
next=df.fillna(method='bfill')
print(prev)
print(next)
```

## LINEAR REGRESSION -5a

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

m, c, r, p, std_err = stats.linregress(x, y)

z=[m * i + c for i in x]

plt.scatter(x, y)
plt.plot(x, z)
plt.show()
```

## POLYNOMIAL REGRESSION -5b

```python
import numpy as np
import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = np.poly1d(np.polyfit(x, y, 3))

myline = np.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

## FINDING OF OUTLIERS -7

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = np.random.randint(1, 21, size=100)

q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = (data < lower_bound) | (data > upper_bound)

plt.boxplot(data)
plt.title('Boxplot')
plt.show()
```

## 10b 3d plotting

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(10)
y = np.random.rand(10)
z = np.random.rand(10)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c='r', marker='o', label='3D scatter plot')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Scatter Plot Example')
ax.legend()

plt.show()
```
